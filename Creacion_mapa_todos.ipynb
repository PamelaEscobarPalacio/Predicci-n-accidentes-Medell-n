```python
def style_function():
    return {
        'color': 'black',  # Color de las líneas
        'weight': 0.5  # Grosor de las líneas (ajusta este valor)
    }

def style_function2(feature):
    return {
        "fillColor": feature["properties"]["color"],
        "fillOpacity": 0.7,
        "stroke": True,
        "strokeOpacity": 0.2,
        "color": "black",
        "weight": 1,
    }

def crear_mapa_todo(datos,diccionario,color_dict,Grupo,muertes,choques,primeras,segundas,terceras):
  barrios_med = gpd.read_file('Barrios de Medellin/Barrio_Vereda.dbf')
  barrios_med['color'] = 'gray'
  barrios_med['grupo'] = 'No clasificado'
  barrios_med['-1'] = ''
  barrios_med['-2'] = ''
  barrios_med['-3'] = ''
  barrios_med['choques'] = ''
  barrios_med['muertes'] = ''

  for lista, barrios in diccionario.items():
    color = color_dict.get(lista, 'gray')  # Obtiene el color correspondiente a la lista o 'gray' si no se encuentra
    grupo = Grupo.get(lista, 'No clasificado')
    muerte = muertes.get(lista, '')
    choque = choques.get(lista, '')
    primera = primeras.get(lista, '')
    segunda = segundas.get(lista, '')
    tercera = terceras.get(lista,'')
    barrios_med.loc[barrios_med['NOMBRE'].isin(barrios), 'color'] = color
    barrios_med.loc[barrios_med['NOMBRE'].isin(barrios), 'grupo'] = grupo
    barrios_med.loc[barrios_med['NOMBRE'].isin(barrios), 'muertes'] = muerte
    barrios_med.loc[barrios_med['NOMBRE'].isin(barrios), 'choques'] = choque
    barrios_med.loc[barrios_med['NOMBRE'].isin(barrios), '-1'] = primera
    barrios_med.loc[barrios_med['NOMBRE'].isin(barrios), '-2'] = segunda
    barrios_med.loc[barrios_med['NOMBRE'].isin(barrios), '-3'] = tercera

  mapa = folium.Map(width=800, height=400, zoom_start=11, location=[6.27,-75.60])
  folium.TileLayer('openstreetmap').add_to(mapa)
  folium.GeoJson(data = barrios_med,
               name = 'NOMBRE',
               style_function = style_function2,
               popup = folium.GeoJsonPopup(
                  fields = ['NOMBRE','grupo','choques','muertes','-1','-2','-3'],
                  aliases = ['Barrio', 'Grupo','Choques:','Muertos:','-','-','-']
               )
               ).add_to(mapa)


  mapa.save('mapa_todos.html')
  return mapa
```
